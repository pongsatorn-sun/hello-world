```python
# To plot pretty figures
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('TISCO.BK.csv', index_col='Date', parse_dates=True)
```

```python
ax = df['Adj Close'].plot(figsize=(12, 6))
sma20_middle = df['Adj Close'].rolling(window=20).mean()
sma20_upper = df['Adj Close'].rolling(window=20).mean() + (df['Adj Close'].rolling(window=20).std()*2)
sma20_lower = df['Adj Close'].rolling(window=20).mean() - (df['Adj Close'].rolling(window=20).std()*2)
sma20_middle.plot(ax=ax)
sma20_lower.plot(ax=ax)
sma20_upper.plot(ax=ax)
```
